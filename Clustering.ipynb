***Theoretical Questions:***

```python
# Question 1. What is unsupervised learning in the context of machine learning
# Answer 1 --> Unsupervised learning is a type of machine learning where models are trained on unlabelled data to discover patterns, groupings, or structures, like clustering or dimensionality reduction.

# Question 2. How does K-Means clustering algorithm work
# Answer 2 --> K-Means partitions data into 'k' clusters by randomly initializing centroids, assigning data points to the nearest centroid, updating centroids to the mean of their clusters, and repeating until convergence.

# Question 3. Explain the concept of a dendrogram in hierarchical clustering
# Answer 3 --> A dendrogram is a tree-like diagram used in hierarchical clustering to represent how clusters are merged or split at each step. It helps in deciding the number of clusters by 'cutting' the tree.

# Question 4. What is the main difference between K-Means and Hierarchical Clustering
# Answer 4 --> K-Means is a flat, partition-based method requiring a predefined number of clusters, while Hierarchical Clustering builds a nested tree of clusters and doesn’t require specifying the number of clusters in advance.

# Question 5. What are the advantages of DBSCAN over K-Means
# Answer 5 --> DBSCAN does not require specifying the number of clusters, can find clusters of arbitrary shape, and handles noise/outliers better than K-Means, which assumes spherical clusters.

# Question 6. When would you use Silhouette Score in clustering
# Answer 6 --> Silhouette Score is used to evaluate how well a data point fits within its cluster. It helps in selecting the optimal number of clusters and assessing clustering performance.

# Question 7. What are the limitations of Hierarchical Clustering
# Answer 7 --> Hierarchical Clustering is computationally expensive for large datasets and once clusters are merged or split, they cannot be undone (no backtracking).

# Question 8. Why is feature scaling important in clustering algorithms like K-Means
# Answer 8 --> K-Means relies on distance metrics like Euclidean distance. Without feature scaling, variables with larger scales dominate, leading to biased clustering results.

# Question 9. How does DBSCAN identify noise points
# Answer 9 --> DBSCAN labels points as noise if they are not core points and don’t fall within the neighborhood of any core point (i.e., insufficient neighbors within a defined radius).

# Question 10. Define inertia in the context of K-Means
# Answer 10 --> Inertia is the sum of squared distances between data points and their assigned cluster centers. Lower inertia indicates tighter clusters and better performance.

# Question 11. What is the elbow method in K-Means clustering
# Answer 11 --> The elbow method plots the number of clusters vs. inertia. The 'elbow point', where the inertia starts to decrease slowly, indicates the optimal number of clusters.

# Question 12. Describe the concept of "density" in DBSCAN
# Answer 12 --> In DBSCAN, density refers to the number of points within a given radius (ε). A region is considered dense if it contains at least the minimum number of points (min_samples).

# Question 13. Can hierarchical clustering be used on categorical data
# Answer 13 --> Yes, hierarchical clustering can be applied to categorical data using suitable distance measures like Hamming distance or using methods like Gower distance.

# Question 14. What does a negative Silhouette Score indicate
# Answer 14 --> A negative Silhouette Score indicates that a sample may be assigned to the wrong cluster, as it is closer to another cluster than to the one it's in.

# Question 15. Explain the term "linkage criteria" in hierarchical clustering
# Answer 15 --> Linkage criteria determine how the distance between clusters is calculated. Common types include single, complete, average, and ward linkage.

# Question 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities
# Answer 16 --> K-Means assumes clusters are spherical and of equal size. It struggles with clusters that have different densities or non-globular shapes, leading to incorrect assignments.

# Question 17. What are the core parameters in DBSCAN, and how do they influence clustering
# Answer 17 --> DBSCAN has two core parameters: ε (epsilon - the radius for neighborhood) and min_samples (minimum points to form a dense region). These define core points and influence cluster shape and count.

# Question 18. How does K-Means++ improve upon standard K-Means initialization
# Answer 18 --> K-Means++ selects initial centroids in a smarter way to spread them out, reducing the chances of poor convergence and improving clustering quality.

# Question 19. What is agglomerative clustering
# Answer 19 --> Agglomerative clustering is a bottom-up hierarchical clustering approach where each data point starts as its own cluster, and pairs are merged iteratively based on a linkage criterion.

# Question 20. What makes Silhouette Score a better metric than just inertia for model evaluation
# Answer 20 --> Silhouette Score considers both intra-cluster tightness and inter-cluster separation, offering a more balanced evaluation than inertia, which only measures compactness.
```

```python
 
# Question 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot
# Answer 21 -->
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=4, random_state=42)
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title("K-Means Clustering on make_blobs")
plt.show()

 
```

```python
# Question 22. Load the Iris dataset and use Agglomerative Clustering to group the data into 3 clusters. Display the first 10 predicted labels
# Answer 22 -->
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering

iris = load_iris()
X = iris.data

agg = AgglomerativeClustering(n_clusters=3)
labels = agg.fit_predict(X)
print("First 10 Agglomerative Clustering Labels:", labels[:10])

 
```

```python
# Question 28. Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means
# Answer 28 -->
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

digits = load_digits()
X = PCA(n_components=2).fit_transform(digits.data)

kmeans = KMeans(n_clusters=10, random_state=42)
labels = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='tab10')
plt.title("K-Means Clustering on Digits (PCA Reduced)")
plt.show()
```

```python
# Question 29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart
# Answer 29 -->
from sklearn.metrics import silhouette_score

X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
scores = []
ks = range(2, 6)

for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=0)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    scores.append(score)

plt.bar(ks, scores, color='orange')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for k = 2 to 5")
plt.show()
```

```python
# Question 30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage
# Answer 30 -->
from scipy.cluster.hierarchy import dendrogram, linkage

X = load_iris().data
linked = linkage(X, method='average')

plt.figure(figsize=(10, 5))
dendrogram(linked, truncate_mode='lastp', p=30)
plt.title("Dendrogram (Average Linkage - Iris Dataset)")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

```python
# Question 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries
# Answer 31 -->
from matplotlib.colors import ListedColormap

X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.5, random_state=42)
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.05),
                     np.arange(y_min, y_max, 0.05))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF']))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, cmap='Set1')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black', marker='X')
plt.title("K-Means Decision Boundaries")
plt.show()
```

```python
# Question 34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot
# Answer 34 -->
X = load_breast_cancer().data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

inertias = []
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.plot(range(2, 7), inertias, marker='o')
plt.title("Inertia vs. Number of Clusters (Breast Cancer Dataset)")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```
